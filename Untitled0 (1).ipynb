```python
import pandas as pd

fake = pd.read_csv("/content/fakenews/Fake.csv")
real = pd.read_csv("/content/fakenews/True.csv")

# Add labels
fake["label"] = 0
real["label"] = 1

# Combine and shuffle
df = pd.concat([fake, real]).sample(frac=1, random_state=42).reset_index(drop=True)
df["content"] = df["title"] + " " + df["text"]

print("Fake:", len(fake), "| Real:", len(real))
print(df["label"].value_counts())
print(df["content"].iloc[0][:300])
```

```python
from transformers import BertTokenizer
from sklearn.model_selection import train_test_split
import torch

tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")

# Trim to 512 tokens max (BERT's limit)
def tokenize(texts, max_len=128):
    return tokenizer(
        list(texts),
        padding=True,
        truncation=True,
        max_length=max_len,
        return_tensors="pt"
    )

# Use a smaller sample for quick training
df = df.sample(5000, random_state=42).reset_index(drop=True)

X_train, X_test, y_train, y_test = train_test_split(
    df["content"], df["label"], test_size=0.2, random_state=42
)

train_enc = tokenize(X_train)
test_enc  = tokenize(X_test)

train_labels = torch.tensor(y_train.values)
test_labels  = torch.tensor(y_test.values)
```

```python
print("Tokenizing train set... (may take ~30 seconds)")
train_enc = tokenize(X_train)

print("Tokenizing test set...")
test_enc = tokenize(X_test)

# Convert labels to tensors
train_labels = torch.tensor(y_train.values)
test_labels  = torch.tensor(y_test.values)

print("Done! Shapes:")
print("Train input IDs:", train_enc["input_ids"].shape)
print("Test input IDs: ", test_enc["input_ids"].shape)
```

```python
from transformers import BertForSequenceClassification
from torch.utils.data import DataLoader, TensorDataset

# Load BERT with a 2-class output head (Fake / Real)
# This will download the model weights ~440MB, takes 1-2 mins
print("Loading BERT model...")
model = BertForSequenceClassification.from_pretrained(
    "bert-base-uncased",
    num_labels=2
)
print("Model loaded!")
```

```python
from torch.utils.data import DataLoader, TensorDataset

# Package data into PyTorch datasets
train_dataset = TensorDataset(
    train_enc["input_ids"],
    train_enc["attention_mask"],
    train_labels
)
test_dataset = TensorDataset(
    test_enc["input_ids"],
    test_enc["attention_mask"],
    test_labels
)

# DataLoaders feed data in batches during training
train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
test_loader  = DataLoader(test_dataset,  batch_size=16)

# Use GPU if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

print("Using device:", device)
print("Train batches:", len(train_loader))
print("Test batches: ", len(test_loader))
```

```python
from torch.optim import AdamW
from sklearn.metrics import accuracy_score, classification_report

optimizer = AdamW(model.parameters(), lr=2e-5)

def train_epoch(loader):
    model.train()
    total_loss = 0
    for batch in loader:
        ids, mask, labels = [b.to(device) for b in batch]
        optimizer.zero_grad()
        outputs = model(input_ids=ids, attention_mask=mask, labels=labels)
        outputs.loss.backward()
        optimizer.step()
        total_loss += outputs.loss.item()
    return total_loss / len(loader)

def evaluate(loader):
    model.eval()
    preds, actuals = [], []
    with torch.no_grad():
        for batch in loader:
            ids, mask, labels = [b.to(device) for b in batch]
            logits = model(input_ids=ids, attention_mask=mask).logits
            preds   += torch.argmax(logits, dim=1).cpu().tolist()
            actuals += labels.cpu().tolist()
    return accuracy_score(actuals, preds), classification_report(actuals, preds, target_names=["Fake","Real"])

print("Functions ready!")
```

```python
# Each epoch takes ~5 mins on GPU
for epoch in range(3):
    print(f"Epoch {epoch+1}/3 training...")
    loss = train_epoch(train_loader)
    print(f"Evaluating...")
    acc, report = evaluate(test_loader)
    print(f"Epoch {epoch+1} | Loss: {loss:.4f} | Accuracy: {acc:.4f}")
    print(report)
    print("-" * 50)
```

```python
def predict(text):
    model.eval()
    enc = tokenizer(
        text,
        return_tensors="pt",
        truncation=True,
        max_length=128,
        padding=True
    ).to(device)

    with torch.no_grad():
        logits = model(**enc).logits
        probs  = torch.softmax(logits, dim=1).cpu().numpy()[0]
        pred   = torch.argmax(logits, dim=1).item()

    label = "✅ REAL" if pred == 1 else "🚨 FAKE"
    print(f"{label}  (Fake: {probs[0]:.2%} | Real: {probs[1]:.2%})")

# Test it out!
predict("WHO confirms new vaccine is safe and effective against latest variant.")
predict("Government secretly putting microchips in drinking water to control citizens.")
predict("NASA scientists discover water ice on the surface of Mars.")
predict("Doctors are hiding the cure for cancer to make more money.")
```

```python
import gradio as gr

def gradio_predict_v2(text):
    if not text.strip():
        return "Please enter some text.", 0.0

    model.eval()
    enc = tokenizer(
        text,
        return_tensors="pt",
        truncation=True,
        max_length=128,
        padding=True
    ).to(device)

    with torch.no_grad():
        logits = model(**enc).logits
        probs  = torch.softmax(logits, dim=1).cpu().numpy()[0]
        pred   = torch.argmax(logits, dim=1).item()

    label      = "✅ REAL NEWS" if pred == 1 else "🚨 FAKE NEWS"
    confidence = float(probs[1] if pred == 1 else probs[0])

    high_fake = "⚠️ High confidence fake — strong misinformation signals detected." if probs[0] > 0.95 else ""
    high_real = "✅ High confidence real — language patterns match credible reporting." if probs[1] > 0.95 else ""
    low_conf  = "🤔 Low confidence — borderline case, treat with caution." if confidence < 0.75 else ""

    explanation = (
        f"{label}  —  {confidence:.2%} confident\n\n"
        f"BERT analysed {len(text.split())} words across 12 transformer\n"
        f"layers and 144 attention heads.\n\n"
        f"Fake probability:  {probs[0]:.2%}\n"
        f"Real probability:  {probs[1]:.2%}\n\n"
        f"{high_fake}{high_real}{low_conf}"
    )

    return explanation, confidence

demo2 = gr.Interface(
    fn=gradio_predict_v2,
    inputs=gr.Textbox(
        lines=5,
        placeholder="Paste a news headline or article here...",
        label="📰 News Text"
    ),
    outputs=[
        gr.Textbox(label="🤖 BERT Analysis", lines=10),
        gr.Number(label="Confidence Score (0 = low, 1 = high)")
    ],
    title="📰 Fake News Detector — Powered by BERT AI",
    description="BERT reads text across 12 transformer layers and 144 attention heads to detect misinformation.",
    examples=[
        ["WHO confirms new vaccine is safe and effective against latest variant."],
        ["Government secretly putting microchips in drinking water to control citizens."],
        ["NASA scientists discover water ice on the surface of Mars."],
        ["Doctors are hiding the cure for cancer to make more money."]
    ],
    theme=gr.themes.Soft()
)

demo2.launch(share=True)
```
